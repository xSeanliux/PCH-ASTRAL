## Duplicate Dataset Rows
Duplicate a dataset's rows given its weighting. For example if a row was weighted 5, it would be split into 5 rows with weight one. Then we can just feed that into ASTRAL.

```python
from pathlib import Path
import pandas as pd
```

```python
RTPOLY = Path("/projects/illinois/eng/cs/warnow/zxliu2/OneMostProb/example/rt_2025_poly")
```

Currently highly weighted characters have weight 10000. That's too high - we will set another number defined below. 

```python
screening_lv = 0
```

```python
def multiply_rows(
    dataset_path: Path,
) -> pd.DataFrame:
    df = pd.read_csv(dataset_path)
    new_rows = []
    for idx, t in df.iterrows():
        w = t.weight
        t.weight = 1
        new_rows.append(pd.DataFrame(
            [t] * w, 
        ))

    res_df = pd.concat(new_rows)
    return res_df
```

```python
for screening_lv in range(4):
    res_df = multiply_rows(RTPOLY / f"rt_2025_poly_screened_lv_{screening_lv}.csv")
    res_df.to_csv(RTPOLY / f"../rt_2025_poly_weighted/rt_2025_poly_screened_lv_{screening_lv}_weighted.csv", index=False)
```
